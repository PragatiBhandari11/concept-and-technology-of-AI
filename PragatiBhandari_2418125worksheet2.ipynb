```python
#QN.2,3
missingvalue = df.isnull().sum()
for col in missingvalue[missingvalue > 0].index:
    if df[col].dtype == 'object':  # For categorical columns
        # Filling with the most frequent value (mode)
        df[col].fillna(df[col].mode()[0], inplace=True)
        print(f"Filled missing values in {col} with mode: {df[col].mode()[0]}")
    else:  # For numeric columns
        # Filling with mean for demonstration; other strategies like median can also be applied
        df[col].fillna(df[col].mean(), inplace=True)
        print(f"Filled missing values in {col} with mean: {df[col].mean()}")
```

```python
#Problem 3.2-data cleaning

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv('/content/drive/MyDrive/concept of AI/Titanic-Dataset.csv')
# Check required columns
required_columns = {'Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived', 'Embarked'}
if not required_columns.issubset(df.columns):
    raise ValueError(f"Dataset missing columns: {required_columns - set(df.columns)}")

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna('S', inplace=True)

# Problem 1: Subset DataFrame and analyze 'Fare' for Pclass == 1
df_subset = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
df_first_class = df_subset[df_subset['Pclass'] == 1]
fare_mean = df_first_class['Fare'].mean()
fare_median = df_first_class['Fare'].median()
fare_max = df_first_class['Fare'].max()
fare_min = df_first_class['Fare'].min()

print("Problem 1:")
print(f"Mean Fare: {fare_mean}")
print(f"Median Fare: {fare_median}")
print(f"Max Fare: {fare_max}")
print(f"Min Fare: {fare_min}")

# Problem 2: Count and drop null values in 'Age'
null_count = df_first_class['Age'].isnull().sum()
df_first_class = df_first_class.dropna(subset=['Age'])

print("\nProblem 2:")
print(f"Number of null values in 'Age': {null_count}")

# Problem 3: One-hot encoding for 'Embarked'
df_encoded = pd.get_dummies(df, columns=['Embarked'], prefix='Embarked', drop_first=False)
print("\nProblem 3:")
print("First few rows of modified DataFrame:")
print(df_encoded.head())

# Problem 4: Survival rates by gender
mean_survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("\nProblem 4:")
print("Mean survival rates by gender:")
print(mean_survival_by_gender)

# Visualization: Survival distributions by gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Problem 5: Survival rates by port of embarkation
mean_survival_by_port = df.groupby('Embarked')['Survived'].mean()
print("\nProblem 5:")
print("Mean survival rates by port of embarkation:")
print(mean_survival_by_port)

# Visualization: Survival rates by port
plt.figure(figsize=(8, 6))
sns.barplot(x='Embarked', y='Survived', data=df, ci=None)
plt.title('Survival Rates by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()

# Problem 6: Survival rates by age group and class
df['AgeGroup'] = pd.qcut(df['Age'], 5, duplicates='drop')
age_class_survival = df.groupby(['AgeGroup', 'Pclass'])['Survived'].mean().unstack()

print("\nProblem 6:")
print("Mean survival rates by age group and passenger class:")
print(age_class_survival)

# Visualization: Survival by age group and class
plt.figure(figsize=(10, 8))
sns.heatmap(age_class_survival, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Survival Rates by Age Group and Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age Group')
plt.show()
```
