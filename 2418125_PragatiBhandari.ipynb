<h2>6. Outlier Detection:</h2>
<h3>• Identify outlier countries in both regions based on Score and GDP per Capita.</h3>
<h3>• Plot these outliers and discuss their implications.</h3>

```python
# Step 1: Define the 1.5 × IQR rule function
def detect_outliers(df, column):
    Q1_score= df[column].quantile(0.25)
    Q3_score = df[column].quantile(0.75)
    IQR = Q3_score - Q1_score
    lower_bound = Q1_score - 1.5 * IQR
    upper_bound = Q3_score + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers for South Asia
sa_outliers_score, sa_lb_score, sa_ub_score = detect_outliers(south_asia_df, 'score')
sa_outliers_gdp, sa_lb_gdp, sa_ub_gdp = detect_outliers(south_asia_df, 'Log GDP per capita')

# Detect outliers for Middle East
me_outliers_score, me_lb_score, me_ub_score = detect_outliers(middle_east_df, 'score')
me_outliers_gdp, me_lb_gdp, me_ub_gdp = detect_outliers(middle_east_df, 'Log GDP per capita')

# Step 2: Combine outliers
sa_outliers = pd.concat([sa_outliers_score, sa_outliers_gdp]).drop_duplicates()
me_outliers = pd.concat([me_outliers_score, me_outliers_gdp]).drop_duplicates()

# Step 3: Scatter plot for South Asia
plt.figure(figsize=(12, 6))
sns.scatterplot(data=south_asia_df, x='Log GDP per capita', y='score', color='blue', label='South Asia')
sns.scatterplot(data=sa_outliers, x='Log GDP per capita', y='score', color='red', label='Outliers', s=100)
plt.title('South Asia: score vs Log GDP per capita with Outliers', fontsize=16)
plt.xlabel('Log GDP per capita', fontsize=14)
plt.ylabel('score', fontsize=14)
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Step 4: Scatter plot for Middle East
plt.figure(figsize=(12, 6))
sns.scatterplot(data=middle_east_df, x='Log GDP per capita', y='score', color='green', label='Middle East')
sns.scatterplot(data=me_outliers, x='Log GDP per capita', y='score', color='red', label='Outliers', s=100)
plt.title('Middle East: score vs Log GDP per capita with Outliers', fontsize=16)
plt.xlabel('Log GDP per capita', fontsize=14)
plt.ylabel('score', fontsize=14)
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Step 5: Display and discuss outliers
print("South Asia Outliers:")
print(sa_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers:")
print(me_outliers[['Country name', 'score', 'Log GDP per capita']])

```
