# 2. Build a Model from Scratch

```python
import numpy as np

# Convert categorical labels to binary (0 = Low, 1 = High)
df_filtered["Heart_Attack_Risk"] = df_filtered["Heart_Attack_Risk"].map({"Low": 0, "High": 1})

# Select features and target variable
features = ["Age", "BMI", "Cholesterol_Level", "Resting_BP", "Heart_Rate", "Max_Heart_Rate_Achieved"]
target = "Heart_Attack_Risk"

X = df_filtered[features].values
y = df_filtered[target].values.reshape(-1, 1)

# Normalize features
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X = (X - X_mean) / X_std

# Add bias term (column of ones for intercept)
X = np.c_[np.ones(X.shape[0]), X]

# Split data manually (80% train, 20% test)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Initialize parameters (weights)
theta = np.zeros((X_train.shape[1], 1))

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Hyperparameters
learning_rate = 0.01
epochs = 1000
m = len(y_train)

# Gradient Descent
for epoch in range(epochs):
    z = X_train @ theta  # Linear combination
    predictions = sigmoid(z)  # Apply sigmoid function
    error = predictions - y_train  # Error term
    gradient = (1/m) * (X_train.T @ error)  # Compute gradient
    theta -= learning_rate * gradient  # Update weights

    if epoch % 100 == 0:  # Print loss every 100 epochs
        loss = -(1/m) * np.sum(y_train * np.log(predictions) + (1 - y_train) * np.log(1 - predictions))
        print(f"Epoch {epoch}: Loss {loss}")

# Predictions on test data
y_pred_prob = sigmoid(X_test @ theta)  # Get probabilities
y_pred = (y_pred_prob >= 0.5).astype(int)  # Convert to binary predictions

# Compute accuracy
accuracy = np.mean(y_pred == y_test) * 100

# Compute Precision, Recall, F1-Score (without sklearn)
TP = np.sum((y_pred == 1) & (y_test == 1))
TN = np.sum((y_pred == 0) & (y_test == 0))
FP = np.sum((y_pred == 1) & (y_test == 0))
FN = np.sum((y_pred == 0) & (y_test == 1))

precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print("\nEvaluation Metrics on Test Data:")
print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```

```python
import numpy as np

# Assume df_filtered is your dataset (already prepared and preprocessed)

# Split features (X) and target (y)
features = ["Age", "BMI", "Cholesterol_Level", "Resting_BP", "Heart_Rate", "Max_Heart_Rate_Achieved"]
target = "Heart_Attack_Risk"

X = df_filtered[features].values
y = df_filtered[target].values.reshape(-1, 1)

# Normalize features (standardize data)
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X = (X - X_mean) / X_std

# Add bias term (column of ones for intercept in linear models)
X = np.c_[np.ones(X.shape[0]), X]

# Split data into training and testing sets (80% train, 20% test)
split_index = int(0.8 * len(X))  # 80% for training
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Now X_train, X_test, y_train, and y_test are ready for use in training and evaluation
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# Define hyperparameter distributions for Logistic Regression
logreg_param_dist = {
    'C': uniform(0.01, 10),  # Regularization strength
    'penalty': ['l1', 'l2'],  # Type of regularization
    'solver': ['liblinear', 'saga']  # Solvers that support L1/L2
}

# Define hyperparameter distributions for Random Forest
rf_param_dist = {
    'n_estimators': randint(50, 300),  # Number of trees
    'max_depth': [None, 10, 20, 30, 40],  # Depth of trees
    'min_samples_split': randint(2, 10),  # Min samples to split
    'min_samples_leaf': randint(1, 5)  # Min samples per leaf
}

# Perform Randomized Search for Logistic Regression
logreg_search = RandomizedSearchCV(
    LogisticRegression(max_iter=1000), logreg_param_dist,
    n_iter=20, cv=5, scoring='f1', random_state=42, n_jobs=-1
)
logreg_search.fit(X_train, y_train.ravel())

# Perform Randomized Search for Random Forest
rf_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42), rf_param_dist,
    n_iter=20, cv=5, scoring='f1', random_state=42, n_jobs=-1
)
rf_search.fit(X_train, y_train.ravel())

# Print best hyperparameters
print("Best Logistic Regression Hyperparameters:", logreg_search.best_params_)
print("Best Random Forest Hyperparameters:", rf_search.best_params_)
```

# 5. Feature Selection

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from scipy.stats import uniform, randint
import pandas as pd
import numpy as np

# Perform Feature Selection
selector = SelectKBest(score_func=f_classif, k=6)  # Select top 6 features
X_train_selected = selector.fit_transform(X_train, y_train.ravel()) # ravel y_train
X_test_selected = selector.transform(X_test)

# Get selected feature names
# Instead of using X_train.columns, use the original features list:
# +1 is added to align with X_train which has a bias column
all_features = ['bias'] + features
selected_features = np.array(all_features)[selector.get_support()]
print("Selected Features:", selected_features)

# Define hyperparameter distributions for Logistic Regression
logreg_param_dist = {
    'C': uniform(0.01, 10),  # Regularization strength
    'penalty': ['l1', 'l2'],  # Type of regularization
    'solver': ['liblinear', 'saga']  # Solvers that support L1/L2
}

# Define hyperparameter distributions for Random Forest
rf_param_dist = {
    'n_estimators': randint(50, 300),  # Number of trees
    'max_depth': [None, 10, 20, 30, 40],  # Depth of trees
    'min_samples_split': randint(2, 10),  # Min samples to split
    'min_samples_leaf': randint(1, 5)  # Min samples per leaf
}

# Perform Randomized Search for Logistic Regression
logreg_search = RandomizedSearchCV(
    LogisticRegression(max_iter=1000), logreg_param_dist,
    n_iter=20, cv=5, scoring='f1', random_state=42, n_jobs=-1
)
logreg_search.fit(X_train_selected, y_train.ravel())

# Perform Randomized Search for Random Forest
rf_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42), rf_param_dist,
    n_iter=20, cv=5, scoring='f1', random_state=42, n_jobs=-1
)
rf_search.fit(X_train_selected, y_train.ravel())

# Print best hyperparameters
print("Best Logistic Regression Hyperparameters:", logreg_search.best_params_)
print("Best Random Forest Hyperparameters:", rf_search.best_params_)
```

The feature selection process using SelectKBest with f_classif is justified for this classification task. It allows us to identify and retain the most relevant features that contribute significantly to predicting heart attack risk. This process enhances the efficiency and effectiveness of our machine learning models, leading to more interpretable and reliable outcomes.

# 6. Final Model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Use the best hyperparameters from Task-4
best_logreg_params = logreg_search.best_params_
best_rf_params = rf_search.best_params_

# Rebuild Logistic Regression Model
logreg_final = LogisticRegression(**best_logreg_params, max_iter=1000)
logreg_final.fit(X_train_selected, y_train.ravel())

# Rebuild Random Forest Model
rf_final = RandomForestClassifier(**best_rf_params, random_state=42)
rf_final.fit(X_train_selected, y_train.ravel())

# Make Predictions
logreg_preds = logreg_final.predict(X_test_selected)
rf_preds = rf_final.predict(X_test_selected)

# Evaluate Models
def evaluate_model(y_true, y_pred, model_name):
    print(f"{model_name} Performance:")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("AUC-ROC:", roc_auc_score(y_true, y_pred))
    print("-")

# Print Evaluation Results
evaluate_model(y_test, logreg_preds, "Logistic Regression")
evaluate_model(y_test, rf_preds, "Random Forest")
```

After rebuilding both the Logistic Regression and Random Forest models using the optimal hyperparameters identified through RandomizedSearchCV (Task-4) and the selected features (Task-5), the performance of the models was evaluated on the test dataset.
 Based on the evaluation, Random Forest appears to be the more robust model for this classification task. It has higher recall and a better overall AUC-ROC score, making it more capable of identifying high-risk individuals (a critical factor in predicting heart attack risk). Random Forest is a powerful ensemble learning method that can capture complex relationships and non-linearities. It is less prone to overfitting due to its averaging technique and generally performs well on diverse datasets. Random Forest is the preferred model for this task, as it handles the complexity of the data better and provides more reliable predictions, especially when the cost of missing high-risk cases is high.
