```python
W = np.array([0,0,0,0])
Y = np.array(hPrice)
print(Y)
```

```python
def cost_function(X, Y, W):
    """ Parameters:
    This function finds the Mean Square Error.
    Input parameters:
      X: Feature Matrix
      Y: Target Matrix
      W: Weight Matrix
    Output Parameters:
      J: accumulated mean square error.
    """
    m = len(Y)

    J = np.sum((X.dot(W) - Y) ** 2)/(2 * m)
    return J
```

```python
initial_cost = cost_function(X2, Y, W)
print(initial_cost)
```

```python
def gradient_descent(X, Y, B, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)

    for iteration in range(iterations):

        Y_pred = X.dot(B)

        loss = Y_pred - Y

        dw = (X.T.dot(loss) ) / (m)

        W_update = W - alpha * dw

        cost = cost_function(X2, Y, W_update)
        cost_history[iteration] = cost

    return W_update, cost_history
```

```python
alpha = 0.0001
new_weights, cost_history = gradient_descent(X2, Y, W, alpha, iterations=1000)


print("New weights:", new_weights)


print("Cost history:", cost_history[-3:])
```

```python
def rmse(Y, Y_pred):
  """
  This Function calculates the Root Mean Squres.
  Input Arguments:
    Y: Array of actual(Target) Dependent Varaibles.
    Y_pred: Array of predeicted Dependent Varaibles.
  Output Arguments:
    rmse: Root Mean Square.
  """
  rmse = np.sqrt(sum((Y-Y_pred)**2)/len(Y))
  return rmse
```

```python
def r2(Y, Y_pred):
  """
   This Function calculates the R Squared Error.
  Input Arguments:
    Y: Array of actual(Target) Dependent Varaibles.
    Y_pred: Array of predeicted Dependent Varaibles.
  Output Arguments:
    rsquared: R Squared Error.
    """
  mean_y = np.mean(Y)
  ss_tot = sum((Y - mean_y) ** 2)
  ss_res = sum((Y - Y_pred) ** 2)
  r2 = 1 - (ss_res / ss_tot)
  return r2
```
