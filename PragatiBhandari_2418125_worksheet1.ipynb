```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n. AB != BA:")
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Are AB and BA equal? ", np.array_equal(AB, BA))
```

```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T = B^T * A^T:")
print("(AB)^T:")
print(AB_transpose)
print("B^T * A^T:")
print(BT_AT)
print("Are (AB)^T and B^T * A^T equal? ", np.array_equal(AB_transpose, BT_AT))
```
